```python
import os
import pandas as pd
import json
import numpy as np
```

```python
first_data = np.load('../data/wordvec.npy')
foodtype = pd.read_csv('../data/food_type.csv')
```

```python
foodtype
```

```python
#numpy 배열을 판다스 데이터프레임으로 변환
first_data = np.squeeze(first_data, axis=0)
df = pd.DataFrame(first_data)
df
```

# 데이터 전처리

```python
food_type = foodtype.to_numpy()[:,1:]
food_type = np.array([np.argmax(v) for v in food_type])
```

```python
foodtype_input = df.to_numpy()
foodtype_target = food_type
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    foodtype_input, foodtype_target, test_size = 0.4, random_state=42)

validate_input, test_input, validate_target, test_target = train_test_split(
    test_input, test_target, test_size = 0.5, random_state=42)
```

#정규화 전처리

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
validate_scaled = ss.transform(validate_input)
test_scaled = ss.transform(test_input)
```

# 소프트 맥스 함수 사용해서 훈련

```python
from sklearn.linear_model import LogisticRegression

lr_sm = LogisticRegression(C=20, max_iter=100)
lr_sm.fit(train_scaled, train_target)


print(lr_sm.score(train_scaled, train_target))
print(lr_sm.score(validate_scaled, validate_target))
```

```python
print(lr_sm.score(test_scaled,test_target))
```

```python
import joblib

#저장된 모델 불러오기
with open('food_type_prediction_model.pkl', 'rb') as model_file:
    loaded_model = joblib.load(model_file)
```

```python
def get_food_type(word_vec):
    with open('food_type_prediction_model.pkl', 'rb') as model_file:
        loaded_model = joblib.load(model_file)

    predicted_labels = loaded_model.predict([word_vec])  # 모델로 예측

    # 예측된 값을 수정하는 로직 추가
    idx = predicted_labels  # 예시로 idx를 1로 설정
    modified_labels = np.zeros(7)
    modified_labels[idx] = 1

    return modified_labels
```
