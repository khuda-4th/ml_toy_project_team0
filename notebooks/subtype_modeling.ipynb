## 로지스틱 회귀

```python
subtype_input = df.to_numpy()
```

```python
print(subtype_input[:5])
```

```python
subtype_target = food_subtype
```

```python
print(subtype_target[:5])
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    subtype_input, subtype_target, test_size = 0.4, random_state=42)
```

```python
validate_input, test_input, validate_target, test_target = train_test_split(
    test_input, test_target, test_size = 0.5, random_state=42)
```

```python
# 정규화 전처리
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
validate_scaled = ss.transform(validate_input)
test_scaled = ss.transform(test_input)
```

```python
# 소프트맥스 분류
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=10, max_iter=100)
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(validate_scaled, validate_target))
```

```python
import pickle

#모델 저장
model = lr

# 모델 저장
with open('food_subtype_m.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

```

```python
lr.score(test_scaled, test_target)
```

```python
print(lr.predict(test_scaled[:20]))
print(test_target[:20])
```

### 하이퍼파라미터 튜닝

```python
from sklearn.linear_model import LogisticRegression

C_list = [0.01,0.1,1,10,30]
max_iter = [10,30,50,80,100,120,140]

for iter in max_iter:
  for c in C_list:
    lr = LogisticRegression(C=c, max_iter=iter)
    lr.fit(train_scaled, train_target)

    print("C:",c," max_iter : ", iter)
    print(lr.score(train_scaled, train_target))
    print(lr.score(validate_scaled, validate_target))

```

## 확률적 경사하강법(SGD Classifier)

```python
from sklearn.linear_model import SGDClassifier
```

```python
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(validate_scaled, validate_target))
```

```python
sc.partial_fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
import numpy as np

sc = SGDClassifier(loss='log_loss', random_state=42)

train_score = []
test_score = []

classes = np.unique(train_target)
```

```python
for i in range(0, 20):
    print(i)
    sc.partial_fit(train_scaled, train_target, classes=classes)

    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

## 결정 트리(Decision Tree)

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=6, random_state=42)
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

## 그레이디언트 부스팅

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_validate

gb = GradientBoostingClassifier(n_estimators=10, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```
